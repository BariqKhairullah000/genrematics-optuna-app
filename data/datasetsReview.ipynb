```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the file path
file_path = 'B:\kuliah\SKRIPSI\Software\genrematics\data/final_combined_movie_data_5genres.csv'

def analyze_movie_genres(file_path):
    """
    Analyze movie genres from a CSV file and create visualizations.
    
    Parameters:
    file_path (str): Path to the CSV file containing movie data
    """
    try:
        # Read the CSV file
        df = pd.read_csv(file_path)
        
        # Count genres
        genre_counts = df['genre'].value_counts()
        total_movies = len(df)
        
        # Print summary
        print("\nGenre Distribution Summary:")
        print("-" * 30)
        for genre, count in genre_counts.items():
            percentage = (count / total_movies) * 100
            print(f"{genre}: {count} movies ({percentage:.1f}%)")
        print("-" * 30)
        print(f"Total Movies: {total_movies}")
        
        # Create visualization
        plt.figure(figsize=(10, 6))
        sns.barplot(x=genre_counts.index, y=genre_counts.values)
        plt.title('Distribution of Movies by Genre')
        plt.xlabel('Genre')
        plt.ylabel('Number of Movies')
        
        # Rotate x-axis labels for better readability
        plt.xticks(rotation=45)
        
        # Add value labels on top of each bar
        for i, v in enumerate(genre_counts.values):
            plt.text(i, v, str(v), ha='center', va='bottom')
        
        # Adjust layout to prevent label cutoff
        plt.tight_layout()
        
        # Show the plot
        plt.show()
        
        return df
        
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
    except Exception as e:
        print(f"Error: An unexpected error occurred: {str(e)}")

# Run the analysis
if __name__ == "__main__":
    df = analyze_movie_genres(file_path)
    
    # Print unique genres in the dataset
    print("\nUnique genres in the dataset:")
    print(df['genre'].unique())
```
